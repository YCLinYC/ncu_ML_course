```python
import os
import shutil
from tqdm import tqdm
import random
```

```python
path = './images'
```

```python
labels = os.listdir(path)
print(labels)
```

```python
os.mkdir('./train')
os.mkdir('./test')
os.mkdir('./validation')
for label in labels:
    os.mkdir('./train/' + label)
    os.mkdir('./test/' + label)
    os.mkdir('./validation/' + label)
    i = 0
    for img in tqdm(os.listdir(path + '/' + label)):
        i = i + 1
        if(i<=100):
            shutil.move(path + '/' + label + '/' + img, './test/' + label + '/' + img)
        elif(i<=200):
            shutil.move(path + '/' + label + '/' + img, './validation/' + label + '/' + img)
        else:
            shutil.move(path + '/' + label + '/' + img, './train/' + label + '/' + img)
```
